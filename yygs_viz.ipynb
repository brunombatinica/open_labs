```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Set random seed for reproducibility
np.random.seed(42)

# Generate data for two normal distributions
n_samples = 10000
mean1, mean2 = 100, 110  # Small mean difference
std1, std2 = 10, 10

# Generate random samples
data1 = np.random.normal(mean1, std1, n_samples)
data2 = np.random.normal(mean2, std2, n_samples)

# Create the plot
plt.figure(figsize=(10, 6))

# Plot density curves
x = np.linspace(-0, 200, 1000)
density1 = norm.pdf(x, mean1, std1)
density2 = norm.pdf(x, mean2, std2)

plt.plot(x, density1, color='blue', linewidth=2, label='No Event', alpha=0.8)
plt.plot(x, density2, color='orange', linewidth=2, label='Event', alpha=0.8)

# Fill areas under curves for better visualization
plt.fill_between(x, density1, alpha=0.3, color='blue')
plt.fill_between(x, density2, alpha=0.3, color='orange')

# Customize the plot
plt.xlabel('Value', fontsize=12)
plt.ylabel('Density', fontsize=12)
plt.title('Two Normal Distributions with Small Mean Difference', fontsize=14)
plt.legend(fontsize=11)
plt.grid(True, alpha=0.3)
plt.xlim(50, 150)
plt.ylim(0, 0.06)

# Add vertical lines for means
plt.axvline(mean1, color='blue', linestyle='--', alpha=0.7, linewidth=1)
plt.axvline(mean2, color='orange', linestyle='--', alpha=0.7, linewidth=1)

plt.tight_layout()
plt.show()

# Print some statistics
print(f"Blue curve (No Event): mean = {mean1}, std = {std1}")
print(f"Orange curve (Event): mean = {mean2}, std = {std2}")
print(f"Mean difference: {abs(mean2 - mean1):.2f}")
```
