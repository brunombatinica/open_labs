```python
import cartopy.crs as ccrs
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
import cartopy.feature as cfeature
from cartopy.io import shapereader
```

```python
# Countries to highlight and their colors
highlighted_countries = {
    'United States': 'lightblue',
    'Brazil':'yellow',
    "United Kingdom":'green',
    "Antarctica":'black',
    # "Russian Federation":'red',
}

# excluded_countries = {
#     'Antarctica',
# }

# Create figure and projection
fig = plt.figure(figsize=(15, 10))
fig.patch.set_facecolor('black') # set area outside of ax to be balck
# ax = plt.axes(projection=ccrs.PlateCarree())
ax = plt.axes(projection=ccrs.Robinson())
ax.set_global()
# ax.coastlines()
ax.add_feature(cfeature.LAND, facecolor='grey', linewidth=0)
ax.add_feature(cfeature.OCEAN, facecolor='black', linewidth=0)
# ax.add_feature(cfeature.BORDERS, linewidth=0.3)

# Load country shapes from Cartopy's Natural Earth data
shpfilename = shapereader.natural_earth(resolution='110m',
                                        category='cultural',
                                        name='admin_0_countries')
reader = shapereader.Reader(shpfilename)
countries = reader.records()

# Draw countries
for country in countries:
    # if country.attributes['NAME_LONG'] in excluded_countries:
    #     continue
    name = country.attributes['NAME_LONG']  # You can also use 'NAME', 'ADM0_A3', etc.
    # print(name)
    color = highlighted_countries.get(name, 'gray') # get else lightgray
    ax.add_geometries(
        [country.geometry], 
        ccrs.PlateCarree(),
        facecolor=color, 
        edgecolor='black', 
        # linewidth=1 if name in highlighted_countries else 0,
        linewidth=0.2,
        alpha=1,
        antialiased=False
    )

# plt.title("World Map Highlighting Specific Countries", fontsize=16)
plt.show()
```
